## 1.importing library

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
import joblib
```

## 2.load dataset

```python
students=pd.read_csv("Student_Performance.csv")
students
```

```python
students.isna().sum()
```

## 3. define X and Y

```python
x=students[["Hours Studied","Previous Scores","Sleep Hours","Sample Question Papers Practiced"]]
x
```

```python
y=students[["Performance Index"]]
y
```

## 4.Split into training and testing sets (80% train, 20% test)

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

## 5. Train the model

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

## 6.Predict on test set

```python
y_predict=model.predict(x_test)
y_predict
```

## 7.Evaluate accuracy

```python
mse = mean_squared_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
```

```python
print("Mean Squared Error (MSE):", mse)
print("R² Score:", r2)
```

## 9. save the trained model

```python
joblib.dump(model,"simple_linear_model.pkl")
print("model saved successfully")
```
